```python
import cv2
from glob import glob
import numpy as np
from numpy import genfromtxt


inputpath = "Tensorflow/workspace/images/spine_dataset/"
outputpath = "Tensorflow/workspace/images/spine_dataset/"
view = "AP"
```

```python

def csvtocoordinate(arr,lumb):
  TL=[0,0]
  TR=[0,0]
  BL=[0,0]
  BR=[0,0]
  arr = arr.astype(float)
  lumbar = {
      "L1":0,
      "L2":2,
      "L3":4,
      "L4":6,
      "L5":8,
  }
  # print(lumbar[lumb])
  TL = [arr[lumbar[lumb]+0][0].item(),arr[lumbar[lumb]+0][1].item()]
  TR = [arr[lumbar[lumb]+0][2].item(),arr[lumbar[lumb]+0][3].item()]
  BL = [arr[lumbar[lumb]+1][0].item(),arr[lumbar[lumb]+1][1].item()]
  BR = [arr[lumbar[lumb]+1][2].item(),arr[lumbar[lumb]+1][3].item()]
  return [int(round(num, 0)) for num in TL],[int(round(num, 0)) for num in TR],[int(round(num, 0)) for num in BL],[int(round(num, 0)) for num in BR]

```

```python

j=0
for i in (glob(inputpath + view+'/*.csv')):
	end = i.find('.csv') 
	filename = i[end-12:end]
	img_path = i.replace("csv", "jpg")
	# print(i)
	img = cv2.imread(img_path)
	height, width, channel = img.shape
	# print(height, width)
	print(j,filename)
	coor = {"L1":[],"L2":[],"L3":[],"L4":[],"L5":[]}
	for k in range(1,6):
		csv = genfromtxt(i, delimiter=',')
		TL,TR,BL,BR = np.round(csvtocoordinate(csv,"L"+str(k))).astype(int)
		# print("L"+str(k),)
		coor["L"+str(k)] = np.array([TL,TR,BL,BR])
	boxcoor = {"L1":{},"L2":{},"L3":{},"L4":{},"L5":{}}
	for i in coor:
	  # print(i)
		xmin = np.min(coor[i].T[0])
		ymin = np.min(coor[i].T[1])
		xmax = np.max(coor[i].T[0])
		ymax = np.max(coor[i].T[1])
		boxcoor[i] ={'xmin':xmin,'ymin':ymin,'xmax':xmax,'ymax':ymax}
  # print(boxcoor['L1']['xmin'])
	text = """<annotation>
    <folder>"""+view+"""</folder>
    <filename>"""+filename +""".jpg</filename>
    <path>D:\\Python\\Tensorflow Object Detection\\TFODCourse\\Tensorflow\\workspace\\images\\collectedimages\\Spine\\"""+view+"""\\"""+filename +""".jpg</path>
    <source>
      <database>Unknown</database>
    </source>
    <size>
      <width>"""+str(width)+"""</width>
      <height>"""+str(height)+"""</height>
      <depth></depth>
    </size>
    <segmented>0</segmented>
    <object>
      <name>"""+view+"""L1</name>
      <pose>Unspecified</pose>
      <truncated>0</truncated>
      <difficult>0</difficult>
      <bndbox>
        <xmin>"""+str(boxcoor['L1']['xmin'])+"""</xmin>
        <ymin>"""+str(boxcoor['L1']['ymin'])+"""</ymin>
        <xmax>"""+str(boxcoor['L1']['xmax'])+"""</xmax>
        <ymax>"""+str(boxcoor['L1']['ymax'])+"""</ymax>
      </bndbox>
    </object>
    <object>
      <name>"""+view+"""L2</name>
      <pose>Unspecified</pose>
      <truncated>0</truncated>
      <difficult>0</difficult>
      <bndbox>
        <xmin>"""+str(boxcoor['L2']['xmin'])+"""</xmin>
        <ymin>"""+str(boxcoor['L2']['ymin'])+"""</ymin>
        <xmax>"""+str(boxcoor['L2']['xmax'])+"""</xmax>
        <ymax>"""+str(boxcoor['L2']['ymax'])+"""</ymax>
      </bndbox>
    </object>
    <object>
      <name>"""+view+"""L3</name>
      <pose>Unspecified</pose>
      <truncated>0</truncated>
      <difficult>0</difficult>
      <bndbox>
        <xmin>"""+str(boxcoor['L3']['xmin'])+"""</xmin>
        <ymin>"""+str(boxcoor['L3']['ymin'])+"""</ymin>
        <xmax>"""+str(boxcoor['L3']['xmax'])+"""</xmax>
        <ymax>"""+str(boxcoor['L3']['ymax'])+"""</ymax>
      </bndbox>
    </object>
    <object>
      <name>"""+view+"""L4</name>
      <pose>Unspecified</pose>
      <truncated>0</truncated>
      <difficult>0</difficult>
      <bndbox>
        <xmin>"""+str(boxcoor['L4']['xmin'])+"""</xmin>
        <ymin>"""+str(boxcoor['L4']['ymin'])+"""</ymin>
        <xmax>"""+str(boxcoor['L4']['xmax'])+"""</xmax>
        <ymax>"""+str(boxcoor['L4']['ymax'])+"""</ymax>
      </bndbox>
    </object>
    <object>
      <name>"""+view+"""L5</name>
      <pose>Unspecified</pose>
      <truncated>0</truncated>
      <difficult>0</difficult>
      <bndbox>
        <xmin>"""+str(boxcoor['L5']['xmin'])+"""</xmin>
        <ymin>"""+str(boxcoor['L5']['ymin'])+"""</ymin>
        <xmax>"""+str(boxcoor['L5']['xmax'])+"""</xmax>
        <ymax>"""+str(boxcoor['L5']['ymax'])+"""</ymax>
      </bndbox>
    </object>
  </annotation>"""
  # print(text)
	f = open(outputpath+view+"/"+filename+".xml", "w")
	f.write(text)
	f.close()
	j=j+1
	# break
# print(filename,boxcoor)
```
